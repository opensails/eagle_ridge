```python
#import Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Request URL
page = requests.get("https://www.pgatour.com/stats/stat.02675.html")
```

```python
#fetch webpage
soup = BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
name = soup.find("td", {"class":"player-name"}).text.replace("\n","").strip()
print(name)
```

```python
names = soup.find_all("td", {"class":"player-name"})
print(name)
```

```python
name_list = []
```

```python
#Iterate through all names


for golfer in range(1):
    all = soup.find_all("td", "player-name")
    for name in all:
        name_list.append(name.text.replace("\n", "").strip())
```

```python
name_list
```

```python
#Data Frame to store scraped data

df = pd.DataFrame({"Names":name_list})

df.to_csv('player_names.csv', index = False)
print(df,"\n Success")
```

```python
stats_list = []
```

```python


for golfer in range(1):
    all = soup.find_all("td", "hidden-small hidden-medium")
    for stat in all:
        stats_list.append(stat.text.replace("\n", "").strip())
```

```python
#Data Frame to store scraped data

df = pd.DataFrame({"Stats":stats_list})

df.to_csv('stats_list.csv', index = False)
print(df,"\n Success")
```
